# imports

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

```

# preprocess

```python
data = pd.read_csv('data/training_set_VU_DM.csv')
```

```python
len(pd.unique(data['prop_id']))
```

```python
for prop in pd.unique(data['prop_id']):
    data.loc[data['prop_id'] == prop, 'booking_prob'] =  data[data['prop_id'] == prop]['booking_bool'].mean()
    data.loc[data['prop_id'] == prop, 'clicking_prob'] =  data[data['prop_id'] == prop]['click_bool'].mean()
```

```python
data['relevance'] = (data['clicking_prob'] + 5 * data['booking_prob'])
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
new_data = pd.DataFrame(columns=data.columns)
for prop in pd.unique(data['prop_id']): 
    new_data.loc[len(new_data)] = data[data['prop_id'] == prop].loc[data[data['prop_id'] == prop].index[0]]
```

```python
new_data
```

```python
data = data.filter(['prop_id','prop_country_id','prop_starrating','prop_review_score','prop_location_score1','relevance'],axis = 1)
data.to_csv('knn_data.csv',index= False)
```

# modeling

```python
data = pd.read_csv('knn_data.csv')
data
```

```python
data = pd.read_csv('knn_data.csv')
data.fillna(0,inplace=True)
labels = data['relevance']
data = data.drop([ 'relevance'],axis = 1)
knn_regressor = KNeighborsRegressor(n_neighbors=1)
transformer = ColumnTransformer([("features", 'passthrough', ['prop_country_id','prop_starrating', 'prop_review_score','prop_location_score1'])])
pipe = Pipeline([('features', transformer), ('knn', knn_regressor)])
pipe.fit(data, labels)
```

# prediction

```python
data_to_predict = pd.read_csv('data/test_set_VU_DM.csv')
data_to_predict.fillna(0,inplace=True)
data_to_predict['predictions'] = pipe.predict(data_to_predict)
data_to_predict = data_to_predict.sort_values(by =['srch_id','predictions'],ascending=[True,False])
submission = data_to_predict.filter(['srch_id','prop_id'],axis = 1)

```

```python
submission.to_csv('submission.csv',index= False)
```
